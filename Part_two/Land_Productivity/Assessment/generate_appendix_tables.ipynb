```python
# 定义输入文件路径
land_cover_path = Path("../data/Land_Cover_Each_Year/C3S_2015_reclass_300m.tif")
productivity_bins = Path("../output/trend/2015/bins.tif")

# 定义土地覆盖类别名称映射（根据实际数据调整）
# 这里是一个示例映射，您需要根据实际的土地覆盖类别数值来调整

land_cover_class_names = {
    1: "Tree-covered areas",
    2: "Grasslands",
    3: "Croplands",
    4: "Wetlands",
    5: "Artificial surfaces",
    6: "Other Lands",
    7: "Water bodies",
}

# 调用函数生成统计表格
df = productivity_land_cover_classification(
    land_cover_path=land_cover_path,
    productivity_bins=productivity_bins,
    land_cover_class_names=land_cover_class_names,  # 可以传入None使用默认数值名称
)

# 显示结果表格
print("Net land productivity trend dynamic (km²) for the baseline period")
print("=" * 80)
os.makedirs("../output/appendix/trend",exist_ok=True)
df.to_csv("../output/appendix/trend/2015_trend.csv")
df
```
